```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_context('paper')
```

# Load a Dataset from Pickle

```python
df = pd.DataFrame(pd.read_pickle('FFI.pkl'),
                  columns=['dataset','name', 'SMILES','image'])
df.head()
```

# Visualize Some Molecules

```python
n_mols = 5
offset = 100

fig, axs = plt.subplots(1,5, figsize=(12,4), constrained_layout=True)

for i, (index, mols) in enumerate(df[offset:offset+n_mols].iterrows()):
    axs[i].set_title(mols[1])
    axs[i].imshow(mols[3])
    sns.despine()
```
