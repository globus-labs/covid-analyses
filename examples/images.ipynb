# Working with Image Files

The examples here require 

* `rdkit`, `pandas`, `seaborn`, and `matplotlib`
* an image file (here `FFI-img.pkl`) downloaded from https://2019-ncovgroup.github.io/data/

```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_context('paper')
```

# Load a Dataset from Pickle

```python
data_file = 'FFI-img.pkl'
df = pd.DataFrame(pd.read_pickle(data_file),
                  columns=['dataset','name', 'SMILES','image'])
df.head()
```

# Visualize Some Molecules

```python
n_mols, offset = (7, 100)

fig, axs = plt.subplots(1, n_mols, figsize=(17,4), constrained_layout=True)
for i, (index, mols) in enumerate(df[offset:offset+n_mols].iterrows()):
    axs[i].set_title(mols[1])
    axs[i].imshow(mols[3])
    sns.despine()
```
